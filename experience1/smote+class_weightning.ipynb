```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, LSTM, Dense, Dropout, BatchNormalization, Add
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r"C:\Users\MSI\Desktop\Mitacs Project\Human Activity Recognition\HAR-WISDM\Data_WISDM\WISDM_cleaned.csv")

# Separate features and target
X = data.drop('class', axis=1)
y = data['class']

# Encode the class labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y_encoded)

# Reshape data for CNN-LSTM (samples, time_steps=10, features=3)
X_resampled = X_resampled.values.reshape(-1, 10, 3)

# Compute class weights
class_weights = compute_class_weight('balanced', classes=np.unique(y_encoded), y=y_encoded)
class_weight_dict = dict(enumerate(class_weights))

# Visualize class distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(x=y, hue=y)
plt.title('Original Class Distribution')
plt.xticks(rotation=45)
plt.subplot(1, 2, 2)
sns.countplot(x=label_encoder.inverse_transform(y_resampled), hue=label_encoder.inverse_transform(y_resampled))
plt.title('Resampled Class Distribution (SMOTE)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


# Define CNN-LSTM model with residual connections
def build_cnn_lstm_model(input_shape, num_classes):
    inputs = Input(shape=input_shape)
    
    # CNN block 1
    x = Conv1D(64, kernel_size=3, padding='same', activation='relu', kernel_regularizer=l2(0.01))(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    
    # Residual connection
    conv1 = Conv1D(64, kernel_size=3, padding='same', activation='relu', kernel_regularizer=l2(0.01))(x)
    conv1 = BatchNormalization()(conv1)
    x = Add()([x, conv1])  # Residual connection
    x = Dropout(0.3)(x)
    
    # CNN block 2
    x = Conv1D(128, kernel_size=3, padding='same', activation='relu', kernel_regularizer=l2(0.01))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    
    # LSTM layer
    x = LSTM(64, return_sequences=False, kernel_regularizer=l2(0.01))(x)
    x = Dropout(0.4)(x)
    
    # Dense layers
    x = Dense(64, activation='relu', kernel_regularizer=l2(0.01))(x)
    x = Dropout(0.4)(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs, outputs)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Initialize stratified k-fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
fold = 1
f1_scores = []

for train_idx, val_idx in skf.split(X_resampled, y_resampled):
    print(f"\nTraining Fold {fold}")
    X_train, X_val = X_resampled[train_idx], X_resampled[val_idx]
    y_train, y_val = y_resampled[train_idx], y_resampled[val_idx]
    
    # Build and train model
    model = build_cnn_lstm_model(input_shape=(10, 3), num_classes=len(np.unique(y_encoded)))
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
    
    history = model.fit(
        X_train, y_train,
        epochs=50,
        batch_size=32,
        class_weight=class_weight_dict,
        validation_data=(X_val, y_val),
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )
    
    # Evaluate model
    y_pred = model.predict(X_val)
    y_pred_classes = np.argmax(y_pred, axis=1)
    print(f"\nFold {fold} Classification Report:")
    print(classification_report(y_val, y_pred_classes, target_names=label_encoder.classes_))
    
    # Store F1-score (macro average)
    report = classification_report(y_val, y_pred_classes, output_dict=True)
    f1_scores.append(report['macro avg']['f1-score'])
    fold += 1

# Print average F1-score across folds
print(f"\nAverage Macro F1-Score across {skf.n_splits} folds: {np.mean(f1_scores):.4f} ± {np.std(f1_scores):.4f}")

# Save the resampled dataset
resampled_data = pd.DataFrame(X_resampled.reshape(-1, 30), columns=X.columns)
resampled_data['class'] = label_encoder.inverse_transform(y_resampled)
resampled_data.to_csv("WISDM_resampled_smote.csv", index=False)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout, BatchNormalization, Bidirectional
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.callbacks import ReduceLROnPlateau

# ---------------------------
# 1. Data preparation
# ---------------------------

# Assuming X and y are your original numpy arrays

# Reshape X: from (samples, 30) to (samples, 10 timesteps, 3 features)
X = X.reshape((X.shape[0], 10, 3))

# Normalize features (StandardScaler on each feature dimension)
scaler = StandardScaler()
# Reshape to 2D for scaler, then back to 3D
X_reshaped = X.reshape(-1, X.shape[-1])
X_scaled = scaler.fit_transform(X_reshaped)
X_scaled = X_scaled.reshape(X.shape)

# Label encode targets
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y.ravel())
y_categorical = to_categorical(y_encoded)

# ---------------------------
# 2. Data augmentation (jittering)
# ---------------------------

def add_jitter(X, sigma=0.05):
    return X + np.random.normal(loc=0.0, scale=sigma, size=X.shape)

# Augment the training data later after train-test split

# ---------------------------
# 3. Train-test split
# ---------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_categorical, test_size=0.2, random_state=42, stratify=y_encoded
)

# Augment training data by jittering
X_train_augmented = add_jitter(X_train)
y_train_augmented = y_train.copy()

# Combine original and augmented data
X_train_combined = np.vstack((X_train, X_train_augmented))
y_train_combined = np.vstack((y_train, y_train_augmented))

# ---------------------------
# 4. Compute class weights
# ---------------------------

class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_encoded),
    y=np.argmax(y_train_combined, axis=1)
)
class_weights_dict = dict(enumerate(class_weights))

# ---------------------------
# 5. Build the CNN-BiLSTM model
# ---------------------------

model = Sequential([
    Conv1D(64, 3, activation='relu', input_shape=(10, 3)),
    BatchNormalization(),
    Conv1D(64, 3, activation='relu'),
    BatchNormalization(),
    MaxPooling1D(2),
    Dropout(0.4),

    Bidirectional(LSTM(100)),
    Dropout(0.4),

    Dense(100, activation='relu'),
    Dropout(0.4),
    Dense(y_categorical.shape[1], activation='softmax')
])

# ---------------------------
# 6. Compile model
# ---------------------------

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

# ---------------------------
# 7. Setup learning rate scheduler
# ---------------------------

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    verbose=1,
    min_lr=1e-6
)

# ---------------------------
# 8. Train model
# ---------------------------

history = model.fit(
    X_train_combined, y_train_combined,
    epochs=100,
    batch_size=64,
    validation_data=(X_test, y_test),
    class_weight=class_weights_dict,
    callbacks=[reduce_lr]
)

# ---------------------------
# 9. Evaluate on test set
# ---------------------------

test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")
```
